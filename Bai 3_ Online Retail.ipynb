```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("OnlineRetail.csv", encoding = "ISO-8859-1")
data.head()
```

```python
print("Số dòng dữ liệu: ",data.shape[0])
print("Số cột dữ liệu: ",data.shape[1])
```

```python
print("Kiểu dữ liệu của từng thuộc tính")
data.info()
```

```python
data=data[data['Country']=='United Kingdom']
print(data)
```

```python
# Xây dựng bảng Pivot table, với mỗi Số hóa đơn tính trung bình cộng số lượng các mặt hàng theo từng Quốc gia. 
data.pivot_table(values=['Quantity'], index=['InvoiceNo','StockCode'], columns='Country', aggfunc ={'Quantity': np.mean})
```

```python
# Xây dựng bảng Pivot table, với mỗi Số hóa đơn tính trung bình cộng số lượng các mặt hàng theo từng Quốc gia. 
data_pivot_hoadon = pd.pivot_table(data= data, index= ['InvoiceNo'],columns= ['Country'], values= 'Quantity', aggfunc='mean' )
print(data_pivot_hoadon)
```

```python
# Xây dựng bảng Pivot table, với mỗi Khách hàng cho biết số lượng mua hàng lớn nhất và nhỏ nhất theo Kho.

data.pivot_table(values='Quantity', index='StockCode', columns='CustomerID', aggfunc = {min, max})
```

```python
# Xây dựng bảng Pivot table, với mỗi Mã kho tính tổng số lượng các mặt hàng và trung bình cộng giá.
data.pivot_table(values=['Quantity','UnitPrice'], index='StockCode', aggfunc ={'Quantity': np.sum,'UnitPrice': np.mean})
```

```python
# Xây dựng bảng Pivot table cho biết tổng số lượng hàng bán được của mỗi ngày.
data.pivot_table(values=['Quantity'],index='InvoiceDate', aggfunc ={'Quantity': np.sum})
```
